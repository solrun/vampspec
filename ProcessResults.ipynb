```python
import numpy as np
import pandas as pd


# Read result data
resultdata = pd.read_pickle('results240210.pkl')

```

```python
resultdict = resultdata[1]
print(resultdict)
```

```python
# Order by problem name
results = pd.DataFrame.from_dict(resultdict,orient='index')
results = results.sort_index()
results
```

```python
# A dataframe that contains only information about whether a proof was found in a given experiment
provedResults = results.copy()
provedResults = provedResults.applymap(lambda x: x[0] == 'uns')
provedResults['proofs found'] = provedResults.sum(axis='columns')
```

```python
# Combine different runs of same schedule
provedResults['proof found'] = provedResults['proofs found'] > 0
provedResults['newtrainind'] = provedResults['newtrainind_1'] + provedResults['newtrainind_2'] + provedResults['newtrainind_3'] + provedResults['newtrainind_4'] + provedResults['newtrainind_5']
provedResults['newtrainind_lemmas'] = provedResults['newtrainind-withlemmas_1'] + provedResults['newtrainind-withlemmas_2'] + provedResults['newtrainind-withlemmas_3'] + provedResults['newtrainind-withlemmas_4'] + provedResults['newtrainind-withlemmas_5']
provedResults['lemmatrainind'] = provedResults['trainind'] + provedResults['trainind_1'] + provedResults['trainind_2'] + provedResults['trainind_3'] + provedResults['trainind_4']
provedResults['lemmatrainind_lemmas'] = provedResults['trainind-withlemmas'] + provedResults['trainind-withlemmas_1'] + provedResults['trainind-withlemmas_2'] + provedResults['trainind-withlemmas_3'] + provedResults['trainind-withlemmas_4']
```

```python
# Some summary statistic sums
provedResults['proof found new schedule'] = provedResults['newtrainind'] + provedResults['newtrainind_lemmas']
provedResults['proof found vanilla'] = provedResults['vanind'] + provedResults['vanind-withlemmas']
provedResults['proof found lemma schedule'] = provedResults['lemmatrainind'] + provedResults['lemmatrainind_lemmas']
provedResults['lemmas needed new schedule'] = provedResults['newtrainind_lemmas'] & (~ provedResults['newtrainind'])
provedResults['lemmas needed vanilla'] = provedResults['vanind-withlemmas'] & (~ provedResults['vanind'])
provedResults['lemmas hinder new schedule'] = (~ provedResults['newtrainind_lemmas']) & provedResults['newtrainind']
provedResults['lemmas hinder vanilla'] = provedResults['vanind'] & (~ provedResults['vanind-withlemmas'])
provedResults['strategy beats vanilla'] = provedResults['proof found new schedule'] & (~ provedResults['proof found vanilla'])
provedResults['vanilla beats strategy'] = provedResults['proof found vanilla'] & (~ provedResults['proof found new schedule'])
provedResults['lemmas needed lemma schedule'] = provedResults['lemmatrainind_lemmas'] & (~ provedResults['lemmatrainind'])
provedResults['lemmas hinder lemma schedule'] = (~ provedResults['lemmatrainind_lemmas']) & provedResults['lemmatrainind']
provedResults.sum()
```

```python
# Statistics for overlap of different methods
provedResults['v without s'] = provedResults['vanind'] & (~ provedResults['newtrainind'])
provedResults['v without s+l'] = provedResults['vanind'] & (~ provedResults['newtrainind_lemmas'])
provedResults['v without ls'] = provedResults['vanind'] & (~ provedResults['lemmatrainind'])
provedResults['v without ls+l'] = provedResults['vanind'] & (~ provedResults['lemmatrainind_lemmas'])
provedResults['v+l without s'] = provedResults['vanind-withlemmas'] & (~ provedResults['newtrainind'])
provedResults['v+l without s+l'] = provedResults['vanind-withlemmas'] & (~ provedResults['newtrainind_lemmas'])
provedResults['v+l without ls'] = provedResults['vanind-withlemmas'] & (~ provedResults['lemmatrainind'])
provedResults['v+l without ls+l'] = provedResults['vanind-withlemmas'] & (~ provedResults['lemmatrainind_lemmas'])
provedResults['s without v'] = provedResults['newtrainind'] & (~ provedResults['vanind'])
provedResults['s without v+l'] = provedResults['newtrainind'] & (~ provedResults['vanind-withlemmas'])
provedResults['s without ls'] = provedResults['newtrainind'] & (~ provedResults['lemmatrainind'])
provedResults['s without ls+l'] = provedResults['newtrainind'] & (~ provedResults['lemmatrainind_lemmas'])
provedResults['s+l without v'] = provedResults['newtrainind_lemmas'] & (~ provedResults['vanind'])
provedResults['s+l without v+l'] = provedResults['newtrainind_lemmas'] & (~ provedResults['vanind-withlemmas'])
provedResults['s+l without ls'] = provedResults['newtrainind_lemmas'] & (~ provedResults['lemmatrainind'])
provedResults['s+l without ls+l'] = provedResults['newtrainind_lemmas'] & (~ provedResults['lemmatrainind_lemmas'])
provedResults['ls without v'] = provedResults['lemmatrainind'] & (~ provedResults['vanind'])
provedResults['ls without v+l'] = provedResults['lemmatrainind'] & (~ provedResults['vanind-withlemmas'])
provedResults['ls without s'] = provedResults['lemmatrainind'] & (~ provedResults['newtrainind'])
provedResults['ls without s+l'] = provedResults['lemmatrainind'] & (~ provedResults['newtrainind_lemmas'])
provedResults['ls+l without v'] = provedResults['lemmatrainind_lemmas'] & (~ provedResults['vanind'])
provedResults['ls+l without v+l'] = provedResults['lemmatrainind_lemmas'] & (~ provedResults['vanind-withlemmas'])
provedResults['ls+l without s'] = provedResults['lemmatrainind_lemmas'] & (~ provedResults['newtrainind'])
provedResults['ls+l without s+l'] = provedResults['lemmatrainind_lemmas'] & (~ provedResults['newtrainind_lemmas'])
provedResults.sum()['v without s':]
```
